```python
import pandas as pd
import numpy as np
```

```python
latlong= pd.read_excel("Grids_LatLong_All.xlsx")
latlong
```

```python
DMA1 = pd.read_excel("DMA1_Data_01012009_02012019.xlsx")
DMA1
```

```python
latlong.drop(columns=['Grid_txt'], inplace=True)
latlong
```

```python
DMA1ll=DMA1.merge(latlong, left_on='1MileGrid', right_on='Grid_no',how='left')
DMA1ll.drop(columns=['Grid_no','Deer ID', 'County', 'Sample Source', '1MileGrid'], inplace=True)
DMA1ll
```

```python
d = {'Valid Test-CWD Not Detected': 0, 'Valid Test-CWD Detected': 1}
DMA1ll.Status = DMA1ll.Status.map(d).fillna(DMA1ll.Status)
e = {'Female': 0, 'Male': 1}
DMA1ll.Sex = DMA1ll.Sex.map(e).fillna(DMA1ll.Sex)
DMA1ll
```

```python
DMA1ll.Sex.value_counts()
```

```python
DMA1ll=DMA1ll[DMA1ll['Sex'] != 'Unknown']
```

```python
DMA1ll["Sample"] = DMA1ll["Sample"].astype('category')
x_with_dummies = pd.get_dummies(DMA1ll["Sample"], sparse=True)
x_final = pd.concat([DMA1ll, x_with_dummies], axis = 1)
del x_final["Sample"]
x_final
```

```python
x_final.to_csv("DMA1_withlatlong.csv", index = False)
```
